```python
# Q2
# Ans:
'''In Bokeh, glyphs are the basic visual building blocks used to create plots. Glyphs represent individual data points or groups
 of data points and are used to create various types of visualizations, such as scatter plots, line charts, bar charts, and more.
 Each glyph corresponds to a specific visual representation, like circles, lines, bars, and patches.

To add glyphs to a Bokeh plot, you typically create a figure object using the figure() function and then call the specific glyph
methods on that figure object. Each glyph method allows you to customize various attributes such as data source, size, color, and
more.
'''


from bokeh.plotting import figure, output_file, show
from bokeh.models import ColumnDataSource

# Sample data
x = [1, 2, 3, 4, 5]
y = [6, 7, 2, 4, 5]

# Specify output file (optional)
output_file("scatter_with_lines.html")

# Create a ColumnDataSource to hold the data
source = ColumnDataSource(data={'x': x, 'y': y})

# Create a figure object
p = figure(title="Scatter Plot with Lines", x_axis_label="X-axis", y_axis_label="Y-axis")

# Add circles glyph
p.circle('x', 'y', size=10, color="blue", alpha=0.5, source=source)

# Add lines glyph
p.line('x', 'y', line_width=2, color="green", source=source)

# Display the plot
show(p)
```

```python
# Q3
# Ans:

# For axis and title and legend
p.title.text = "Customized Title"
p.xaxis.axis_label = "Custom X-axis Label"
p.yaxis.axis_label = "Custom Y-axis Label"
legend_label = 'red points'
show(p)
```
