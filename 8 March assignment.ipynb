```python
# Ans:
The Probability Density Function(PDF) defines the probability function representing the density of a continuous
random variable lying between a specific range of values. In other words, the probability density function
produces the likelihood of values of the continuous random variable. Sometimes it is also called a probability
distribution function or just a probability function.

In simple terms, the PDF helps you understand how likely different outcomes are within a continuous range. 
It's important to note that the PDF itself doesn't directly give you the probability of a specific value occurring, 
but it tells you how the probabilities are distributed over the entire range.
```

```python
# Ans:
Types of Probabilty distributions are as follow:
  
1. Normal (Gaussian) Distribution: One of the most well-known distributions, often seen in natural phenomena,
   characterized by its bell-shaped curve.
    
2. Bernoulli Distribution: Models a binary outcome (success/failure) with a single trial.

3. Uniform Distribution: All values in a given interval are equally likely to occur.

4. Binomial Distribution: Models the number of successes in a fixed number of independent Bernoulli trials.

5. Poisson Distribution: Models the number of events occurring in a fixed interval of time or space, given a certain 
   average rate.

6. Log-Normal Distribution: Often used to model variables that are the result of multiplying many small independent 
  factors.
```

```python
from scipy.stats import norm

def normal_pdf(x, mean, std_dev):
  
    pdf_value = norm.pdf(x, loc=mean, scale=std_dev)
    return pdf_value

# Example usage
mean = 0
std_dev = 1
x_value = 1.5
pdf_at_x = normal_pdf(x_value, mean, std_dev)
print(f"PDF at x = {x_value}: {pdf_at_x}")

```

```python
# Ans:
import numpy as np
import matplotlib.pyplot as plt

# Parameters for the binomial distribution
n = 10  # Number of trials
p = 0.4  # Probability of success

# Generate a random sample of size 1000 from the binomial distribution
sample = np.random.binomial(n, p, size=1000)

# Plot a histogram
plt.hist(sample, bins=np.arange(0, n+2)-0.5, density=True, edgecolor='black')
plt.xlabel('Number of Successes')
plt.ylabel('Probability')
plt.title(f'Binomial Distribution (n={n}, p={p})')
plt.xticks(range(n+1))
plt.grid(True)
plt.show()
```

```python
# ANs:
from scipy.stats import poisson

def poisson_cdf(x, mean):
   
    cdf_value = poisson.cdf(x, mu=mean)
    return cdf_value

# Example usage
mean = 3
x_value = 4
cdf_at_x = poisson_cdf(x_value, mean)
print(f"CDF at x = {x_value}: {cdf_at_x}")
```

```python
# Ans:
The Binomial distribution and the Poisson distribution are both probability distributions used to model discrete events, 
but they are typically applied in different situations and have distinct characteristics. Here are the key differences 
between them:

1. **Nature of Events:**
   - **Binomial Distribution:** The Binomial distribution is used when you have a fixed number of independent trials,each 
       with two possible outcomes (success or failure). The probability of success remains constant across all trials.
   - **Poisson Distribution:** The Poisson distribution is used when you are interested in modeling the number of events 
       that occur in a fixed interval of time or space. It's often applied to situations where events happen randomly and 
       independently.

2. **Number of Trials:**
   - **Binomial Distribution:** Requires a fixed number of trials (denoted as 'n') to be specified in advance.
   - **Poisson Distribution:** Does not involve a fixed number of trials; it's about counting events that occur over a fixed
       interval.

3. **Outcomes per Trial:**
   - **Binomial Distribution:** Each trial has two possible outcomes: success or failure.
   - **Poisson Distribution:** The Poisson distribution models the count of events occurring in a continuous range, so 
       there's no notion of success/failure like in the Binomial distribution.

4. **Parameters:**
   - **Binomial Distribution:** The parameters are 'n' (number of trials) and 'p' (probability of success in each trial).
   - **Poisson Distribution:** The parameter is 'lambda' (average rate of events occurring in the interval).

5. **Assumptions:**
   - **Binomial Distribution:** Assumes a fixed number of trials, independence between trials, and a constant probability of 
       success.
   - **Poisson Distribution:** Assumes that events occur randomly and independently, and the average rate of events is 
       relatively small in the interval.

6. **Formula and Use Cases:**
   - **Binomial Distribution:** Used for situations like counting the number of successful outcomes in a certain number of 
       coin flips, or counting the number of defective items in a production batch.
   - **Poisson Distribution:** Used for modeling events like the number of phone calls received by a call center in an hour,
       or the number of cars passing through a toll booth in a given time.

In summary, the Binomial distribution is appropriate when dealing with a fixed number of trials and two possible outcomes per
trial, while the Poisson distribution is suitable for modeling the count of rare events that occur randomly and independently
in a fixed interval.
```

```python
# Ans:
import numpy as np

# Parameters for the Poisson distribution
mean = 5

# Generate a random sample of size 1000 
sample = np.random.poisson(mean, size=1000)

# Calculate the sample mean and variance
sample_mean = np.mean(sample)
sample_variance = np.var(sample)

print(f"Sample Mean: {sample_mean}")
print(f"Sample Variance: {sample_variance}")
```
