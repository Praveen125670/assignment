```python
# Ans:
Types of Statistics:
    
Basically, there are two types of statistics.

1.Descriptive Statistics
2.Inferential Statistics

Explanation:
 Descriptive Statistics:

The data is summarised and explained in descriptive statistics. The summarization is done from a population sample 
utilising several factors such as mean and standard deviation. Descriptive statistics is a way of organising, 
representing, and explaining a set of data using charts, graphs, and summary measures. Histograms, pie charts, bars, and
scatter plots are common ways to summarise data and present it in tables or graphs. Descriptive statistics are just that:
descriptive. They don’t need to be normalised beyond the data they collect.

 Inferential Statistics:

We attempt to interpret the meaning of descriptive statistics using inferential statistics. We utilise inferential 
statistics to convey the meaning of the collected data after it has been collected, evaluated, and summarised. The
probability principle is used in inferential statistics to determine if patterns found in a study sample may be 
extrapolated to the wider population from which the sample was drawn. Inferential statistics are used to test hypotheses
and study correlations between variables, and they can also be used to predict population sizes. Inferential statistics 
are used to derive conclusions and inferences from samples, i.e. to create accurate generalisations.
```

```python
# Ans
    The concept of levels of measurement, also known as scales of measurement or data types, refers to the different 
    ways in which data can be categorized based on the nature of the values they represent. These levels determine the 
    kinds of statistical analyses and operations that can be applied to the data. There are four primary levels of 
    measurement:
      nominal, ordinal, interval, and ratio.

(i) Nominal Level:
At the nominal level, data are categorized into distinct categories or groups with no inherent order or ranking. 
The data are simply labeled or named. Nominal data is qualitative and does not allow for any mathematical operations
beyond counting or frequency calculations.

Example: Eye color of individuals (blue, brown, green, etc.). Here, there is no inherent order or ranking among the 
colors;they are just distinct categories.

(ii) Ordinal Level:
Ordinal data involves categories that have a meaningful order or ranking, but the differences between them are not
necessarily consistent or measurable. While you can establish a relative ranking, you can't determine precise intervals
between values.

Example: Educational attainment levels (high school, college, graduate). While there's a clear order, the difference 
between "high school" and "college" might not be the same as between "college" and "graduate."

(iii) Interval Level:
Interval data have a meaningful order, and the intervals between values are consistent and measurable. However, 
interval data lack a true zero point, and operations like multiplication and division don't hold meaningful 
interpretations.

Example: Temperature in Celsius or Fahrenheit. While you can say 20°C is hotter than 10°C, you can't say that 20°C is 
twice as hot as 10°C.

(iv) Ratio Level:
Ratio data also have a meaningful order, consistent intervals, and a true zero point. This allows for meaningful 
multiplication and division operations. Ratio data are the most mathematically sophisticated level.

Example: Height, weight, age (in years), and the number of mangoes in a basket. A height of 180 cm is not only taller 
than 160 cm but also represents twice as much height. 
```

```python
# Ans:Descriptive Statistics:
Descriptive statistics involve methods used to summarize, organize, and describe the main features of a dataset. They 
provide a clear and concise overview of the data without making any inferences or drawing conclusions beyond what is 
presented in the data itself. Descriptive statistics are useful for gaining a preliminary understanding of the data and
for communicating its key characteristics.

Example: Let's consider a dataset of exam scores for a class of students. Descriptive statistics for this dataset might
         include the mean (average) score, the median (middle score), the standard deviation (measure of spread), and 
        the minimum and maximum scores. These statistics provide a summary of the central tendency, variability, and the
        range of scores in the class.

Inferential Statistics:
Inferential statistics involve methods used to make inferences or draw conclusions about a population based on a sample
of data. These methods allow us to generalize findings from the sample to the larger population from which the sample was
drawn. Inferential statistics use probability theory to estimate parameters, test hypotheses, and make predictions. They
help us make statements about the larger context using information from a smaller subset of data.

Example: Imagine you're a researcher and you want to determine whether a new teaching method has led to a significant 
         improvement in exam scores compared to the traditional method. You could randomly select a sample of students 
         from a population and teach them using the new method. After the exam, you use inferential statistics to analyze
         the data and determine if the observed difference in scores between the two methods is statistically significant.
         If it is, you can infer that the new teaching method is likely to have an impact on the larger population of
         students.
            
In summary:

(i) Descriptive statistics provide a summary of the characteristics of a dataset, helping us understand its basic features.

(ii) Inferential statistics allow us to make broader conclusions about a population based on a sample, using probability 
     and statistical techniques.
```

```python
# Ans: 
In Mathematics, statistics are used to describe the central tendencies of the grouped and ungrouped data. The three 
measures of central tendency are:
    
1. Mean: The mean, also known as the average, is calculated by summing up all the values in a dataset and dividing by the
   total number of values. It's sensitive to extreme values and can be affected by outliers. The mean is used to represent
    the typical value of a dataset.

2. Median: The median is the middle value in a dataset when it's arranged in order. It's not affected by extreme values or
   outliers as much as the mean. The median is used when the data is skewed or contains outliers that could bias the mean.

3. Mode: The mode is the value that appears most frequently in a dataset. It's used to identify the most common value or 
   values in a dataset. A dataset can have no mode, one mode (unimodal), or multiple modes (multimodal).

Measures of Variability:
Measures of variability quantify the spread or dispersion of data points in a dataset. They provide information about how 
much the individual values deviate from the central value. Common measures of variability include:

1. Range: The range is the difference between the maximum and minimum values in a dataset. It gives an idea of the total
   spread of the data. However, it's highly influenced by extreme values and doesn't provide a detailed picture of 
    variability.

2. Variance: Variance measures the average squared deviation of each data point from the mean. It provides a measure of 
   the overall dispersion of the data. A larger variance indicates greater variability.

3. Standard Deviation: The standard deviation is the square root of the variance. It's a widely used measure of 
   variability that indicates how much individual data points deviate from the mean. A smaller standard deviation implies
    less dispersion, while a larger one suggests greater variability.

4. Interquartile Range (IQR): The IQR is the difference between the third quartile (Q3) and the first quartile (Q1). It 
   represents the spread of the middle 50% of the data and is less influenced by extreme values than the range.
```
