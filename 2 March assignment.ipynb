```python
# Ans:
    Matplotlib is a popular Python library for creating static, interactive, and for visualizations. 
    It provides a wide range of functions to generate various types of plots and charts, making it a 
    powerful tool for data visualization and exploration.
    
    Five plots are as follow:
    
    1.Line Plot: 
    Used to display data points connected by straight lines, making it suitable for
    representing trends or patterns in sequential data.

    2.Scatter Plot: 
    Used to visualize the relationship between two continuous variables, with data 
    points plotted individually, allowing for the identification of patterns or correlations.
    
    3.Bar Plot: 
    Suitable for displaying categorical data with rectangular bars, where the length of
    each bar represents the value of the corresponding category.

    4.Histogram: 
    Used to represent the distribution of a continuous variable. It divides the data 
    into intervals (bins) and displays the frequency or count of data points falling into each bin.

    5.Pie Chart: 
    A circular chart that divides a data set into slices, where each slice represents 
    a proportion of the whole data set. It is useful for showing the composition of a categorical 
    variable in a visually intuitive way.
```

```python
# Ans:
import numpy as np
import matplotlib.pyplot as plt

# Data
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
colours = np.random.rand(50)

# Plotting graph
plt.scatter(x,y,c=colours)
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("Random dataset")

# Display the graph
plt.show()
```

```python
# Ans:
The subplot() function in Matplotlib is used to create a grid of subplots within a single figure. 
It allows you to display multiple plots in a structured layout, making it easier to compare and 
analyze different aspects of data or visualize various datasets side by side. The subplot() 
function is particularly useful when you want to create complex plots containing multiple subplots, 
such as a grid of charts, histograms, or scatter plots.

The general syntax of the subplot() function is:
    
matplotlib.pyplot.subplot(rows, columns, index)
```

```python
import numpy as np
import matplotlib.pyplot as plt

#For line 1: 
x = np.array([0, 1, 2, 3, 4, 5]) 
y = np.array([0, 100, 200, 300, 400, 500])
plt.subplot(2,2,1)
plt.plot(x,y)
plt.title("For line 1")


#For line 2: 
x = np.array([0, 1, 2, 3, 4, 5]) 
y = np.array([50, 20, 40, 20, 60, 70])
plt.subplot(2,2,2)
plt.plot(x,y)
plt.title("For line 2")

#For line 3: 
x = np.array([0, 1, 2, 3, 4, 5]) 
y = np.array([10, 20, 30, 40, 50, 60])
plt.subplot(2,2,3)
plt.plot(x,y)
plt.title("For line 3")

#For line 4: 
x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([200, 350, 250, 550, 450, 150])
plt.subplot(2,2,4)
plt.plot(x,y)
plt.title("For line 4")

# Adjusting layout to prevent overlapping titles and labels
plt.tight_layout()

# Displaying the entire figure with all subplots
plt.show()
```

```python
# Ans: 
Bar Plot: 
    A bar plot is a type of chart used in data visualization to represent categorical data with rectangular bars.
    where the length of each bar represents the value of the corresponding category.
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Data
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

# Bar graph
plt.subplot(2,1,1)
plt.bar(company,profit)
plt.xlabel("name of companies")
plt.ylabel("data")
plt.title("company data")

# Horizontal bar graph
plt.subplot(2,1,2)
plt.barh(company,profit)
plt.ylabel("name of companies")
plt.xlabel("data")
plt.title("horizontal company data")

# Adjusting layout to prevent overlapping titles and labels
plt.tight_layout()

# Displaying the entire figure with all subplots
plt.show()
```

```python
# Ans:
A box plot, also known as a box-and-whisker plot, is a statistical chart used to represent the distribution
of a continuous dataset. It provides a concise summary of the data's central tendency, dispersion, 
and skewness, making it useful for visualizing the spread and identifying outliers in the data.
```

```python
import matplotlib.pyplot as plt

# Sample data
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

# Creating a box plot
plt.subplot(2,1,1)
plt.boxplot(box1)
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("title")

# Creating a box plot
plt.subplot(2,1,2)
plt.boxplot(box2)
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("title")

# Adjusting layout to prevent overlapping titles and labels
plt.tight_layout()

# Displaying the entire figure with all subplots
plt.show()
```
