```python
Where: ρ is the Spearman's rank correlation coefficient.
     d represents the differences between the ranks.
n is the number of data points (individuals in this case).

Interpret the result:
If ρ=1, it indicates a perfect monotonic positive relationship, meaning that as the amount of sleep individuals
get increases, their job satisfaction also consistently increases.

If ρ=−1, it indicates a perfect monotonic negative relationship, meaning that as the amount of sleep individuals
get increases, their job satisfaction consistently decreases.

If ρ=0, it suggests no monotonic relationship between the two variables.

Interpreting the result also depends on the magnitude of ρ. Larger absolute values of ρ indicate stronger
monotonic relationships.

Remember that Spearman's rank correlation measures monotonicity but does not assume linearity, so it can
capture relationships that Pearson's correlation might miss.
```

```python
df=pd.read_csv('exercise_dataset.csv')
```

```python
df2=df[['Exercise Intensity','BMI']]
```

```python
df2.corr(method='spearman')
```

```python
df2.corr(method='pearson')
```

```python
Ans:
Interpret the result:
If r is close to 1, it indicates a strong positive linear relationship between exercise hours and BMI.
If r is close to -1, it indicates a strong negative linear relationship.
If r is close to 0, it suggests a weak or no linear relationship.

Interpret the result:
Similar to the interpretation of Pearson's correlation, ρ close to 1 indicates a strong positive monotonic
relationship, ρ close to -1 indicates a strong negative monotonic relationship, and ρ close to 0 suggests
no monotonic relationship.

Comparison:

If both Pearson and Spearman's correlations are close to 1 or -1, it suggests a strong relationship.
If Pearson is close to 0 but Spearman's is not, it implies a non-linear but monotonic relationship.
If both are close to 0, there may be no clear relationship between the variables.
By calculating and comparing both coefficients, you can gain a better understanding of the relationship 
between exercise hours and BMI in your sample of adults.
```

```python
df=pd.DataFrame({
      'screen time':[8,8,19,6,12,8,4,7],
      'physical activity':[4,3,1,7,4,6,3,4]
  })
```

```python
df.corr(method='pearson')
```

```python
Ans:
    Interpret the result:
If r is close to 1, it indicates a strong positive linear relationship. This means that as the number
of hours spent watching television per day increases, the level of physical activity tends to increase.

If r is close to -1, it indicates a strong negative linear relationship. This means that as the number 
of hours spent watching television per day increases, the level of physical activity tends to decrease.

If r is close to 0, it suggests a weak or no linear relationship between the two variables. In this case, 
the number of hours spent watching television per day may not have a strong linear influence on the level
of physical activity.

Keep in mind that correlation does not imply causation. The Pearson correlation coefficient will only 
tell you about the strength and direction of a linear relationship between these two variables, not
whether one variable causes the other.
```
