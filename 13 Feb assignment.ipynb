```python
# Q3. What errors are defined in the ArithmeticError class? Explain any two with an example.

# Ans-> Example 1 - Handling ZeroDivisionError
try:
    arithmetic = 5/0
    print(arithmetic)
except ArithmeticError as e:
    print(f'You have just made an Arithmetic error,{e.__class__}')
```

```python
# Example 2 - Handling OverflowError
j = 5.0

try:
    for i in range(1, 1000):
        j = j**i
except ArithmeticError as e:
    print(f"{e},{e.__class__}")
```

```python
# Q4. Why LookupError class is used? Explain with an example KeyError and IndexError.

# Ans-> The LookupError exception in Python forms the base class for all exceptions that are raised when an index 
#       or a key is not found for a sequence or dictionary respectively.

# Example 1 - Handling IndexError exception
x = [1, 2, 3, 4]
try:
    print(x[10])
except LookupError as e:
    print(f"{e}, {e.__class__}")
        
```

```python
# Example 2 - Handling KeyError exception
info = {'name': 'Lenin Mishra',
                'age': 28,
                'language': 'Python'}
l = input('What do you want to learn about Pylenin==> ')

try:
    print(f'{l} is {info[l]}')
except LookupError as e:
    print(f'{e}, {e.__class__}')
```
