```python
Ans->
These are the HTTPS methods.
GET:-
GET is used to request data from a specific resource. The best example is URL, its consist of GET request.
eg: when you are searching on the google.

POST:-
POST is used to send data to a server to create/update a resource. The data sent to the server with POST is stored
in the request body of the HTTP request.
eg:- E-mail login page of the the e_mail site. 
```

```python
Ans-> The Request, in Flask, is an object that contains all the data sent from the Client to Server. 
 This data can be recovered using the GET/POST Methods.
```

```python
Ans-> Flask redirect is defined as a function or utility in Flask which allows developers to redirect users to a specified URL
      and assign a specified status code. When this function is called,a response object is returned, and the redirection happens
      to the target location with the status code.
```
