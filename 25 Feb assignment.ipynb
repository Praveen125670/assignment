# Ans:
loc():
The loc() function is label based data selecting method which means that we have to pass the name of 
the row or column which we want to select. This method includes the last element of the range passed 
in it, unlike iloc(). loc() can accept the boolean data unlike iloc().

iloc():
The iloc() function is an indexed-based selecting method which means that we have to pass an integer 
index in the method to select a specific row/column. This method does not include the last element 
of the range passed in it unlike loc(). iloc() does not accept the boolean data unlike loc().

```python
# Ans:
import pandas as pd

import numpy as np

columns = ['column_1', 'column_2', 'column_3', 'column_4', 'column_5', 'column_6']

indices = [1,2,3,4,5,6]

#Creating a dataframe:

df1 = pd.DataFrame(np.random.rand(6,6), columns = columns, index = indices)

#reindexing the dataframe
new_df = df1.reindex([3,0,1,2])

#printing as Dataframe
pd.DataFrame(new_df)


# printing output
print("new_df output of new_df.loc[2] : \n",new_df.loc[2])
print("\n new_df output of new_df.iloc[2]: \n",new_df.iloc[2])
```

```python
new_df
```

Here in the program new_df.loc[2] will look for the label because the loc() function is label based data selecting method,
in this case label is 2.

And in the case of new_df.iloc[2] will look for the index because the iloc() function is an indexed-based selecting method,
in this case on the 2 position index is 1 because of reindexing

```python
#Ans:
# importing required library
from datetime import datetime

#getting date input from user
start_date = input("Enter a date(YYYY-MM-DD) : ")
end_date = input("Enter a date(YYYY-MM-DD) : ")

# Using strptime to format the time stamp which is in string format to date-time object.
date_1 = datetime.strptime(start_date,"%Y-%m-%d")
date_2 = datetime.strptime(end_date,"%Y-%m-%d")

# getting difference
difference =date_2-date_1

# printing the output
print("Difference in dates are as follow : ")
print("Days : ",difference.days)
print("Hours : ",difference.seconds//3600)
print("Minutes : ",(difference.seconds // 60)%60)
```

```python
# Ans:
import pandas as pd

def read_csv_file(file_path):
    try:
        df = pd.read_csv(file_path)
        return df
    except FileNotFoundError:
        print("File not found. Please provide the correct file path.")
        return None

def convert_to_categorical(df, column_name, category_order):
    try:
        df[column_name] = pd.Categorical(df[column_name], categories=category_order, ordered=True)
        return df
    except KeyError:
        print("Invalid column name. Please provide the correct column name.")
        return None

def main():
    file_path = input("Enter the CSV file path: ")
    df = read_csv_file(file_path)

    if df is not None:
        column_name = input("Enter the column name to convert to categorical: ")

        category_order_str = input("Enter the category order (comma-separated): ")
        category_order = [cat.strip() for cat in category_order_str.split(",")]

        df = convert_to_categorical(df, column_name, category_order)

        if df is not None:
            sorted_df = df.sort_values(by=[column_name])
            print("\nSorted Data:")
            print(sorted_df)

if __name__ == "__main__":
    main()
```

```python
# Ans:
import pandas as pd
import matplotlib.pyplot as plt

def read_csv_file(file_path):
    try:
        df = pd.read_csv(file_path)
        return df
    except FileNotFoundError:
        print("File not found. Please provide the correct file path.")
        return None

def plot_stacked_bar_chart(df):
    if 'Date' in df.columns and 'Product line' in df.columns and 'Total' in df.columns:
        df['Date'] = pd.to_datetime(df['Date'])
        df_pivot = df.pivot_table(index='Date', columns='Product line', values='Total', aggfunc='sum', fill_value=0)
        df_pivot.plot(kind='bar', stacked=True, figsize=(12, 8))
        plt.xlabel('Date')
        plt.ylabel('Total')
        plt.title('Sales of Different Products Over Time')
        plt.legend(loc='upper left')
        plt.show()
    else:
        print("The required columns (Date, Product line, Total) not found in the CSV file.")

def main():
    file_path = input("Enter the CSV file path: ")
    df = read_csv_file(file_path)

    if df is not None:
        plot_stacked_bar_chart(df)

if __name__ == "__main__":
    main()
```
