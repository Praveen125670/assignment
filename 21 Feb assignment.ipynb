```python
Ans->Web Scrapers can be divided on the basis of many different criteria, including Self-built or Pre-built Web Scrapers,
     Browser extension or Software Web Scrapers, and Cloud or Local Web Scrapers.
    
     Self-built Web Scrapers but that requires advanced knowledge of programming.
     Pre-built Web Scrapers are previously created scrapers that you can download and run easily. 
     Browser extensions Web Scrapers are extensions that can be added to your browser.
     Software Web Scrapers don’t have these limitations as they can be downloaded and installed on your computer.
     Cloud Web Scrapers run on the cloud.
     Local Web Scrapers, on the other hand, run on your computer using local resources.  
    
```
