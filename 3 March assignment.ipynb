```python
# Ans:
Seaborn is a powerful Python data visualization library built on top of Matplotlib. It provides a wide range of statistical
plots that help to visualize data effectively. Here are five common plots that you can create using Seaborn, along with their
uses:

Scatter Plot:
Use: Scatter plots are used to visualize the relationship between two numerical variables. They help to identify patterns,
correlations, or trends between the variables.

Line Plot:
Use: Line plots are used to visualize the trend of a numerical variable over a continuous interval. They are useful for 
showing how a variable changes over time or other continuous scales.

Bar Plot:
Use: Bar plots are used to compare categorical data. They are particularly helpful in visualizing the distribution of a 
categorical variable or comparing the values of different categories.

Histogram:
Use: Histograms are used to display the distribution of a continuous numerical variable. They help identify the underlying
data distribution, such as whether it is skewed, symmetric, or bimodal.

Box Plot (Box-and-Whisker Plot):
Use: Box plots are used to visualize the distribution of a numerical variable across different categories. They provide 
information about the median, quartiles, and potential outliers in the data.
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

fmri = sns.load_dataset('fmri')
sns.lineplot(x='timepoint', y='signal', data=fmri)

plt.show()
```

```python
# Ans:
import seaborn as sns
import matplotlib.pyplot as plt

titanic = sns.load_dataset('titanic')
sns.boxplot(x='pclass', y='age',data=titanic)

plt.show()
```

```python
# Ans: 
import seaborn as sns
import matplotlib.pyplot as plt

diamond = sns.load_dataset('diamonds')
sns.histplot(x = 'price',hue = 'cut',data = diamond,bins=40)

plt.show()
```

```python
# Ans:
import seaborn as sns
import matplotlib.pyplot as plt

iris = sns.load_dataset('iris')
sns.pairplot(hue = 'species',data = iris)

plt.show()
```

```python
# Ans:
import seaborn as sns
import matplotlib.pyplot as plt

flights = sns.load_dataset('flights')
pivot = flights.pivot('month','year','passengers')

plt.figure(figsize=(10,7))
sns.heatmap(pivot,annot = True,fmt='d',cmap="YlGnBu")

plt.xlabel("Year")
plt.ylabel("Month")
plt.title("Passenger Flights Heatmap")

plt.show()
```
