```python
# Ans:
import numpy as np

list_ = [ '1' , '2' , '3' , '4' , '5' ]

array_list = np.array(object = list_)

# printig data type of the list
print("Datatype of list is : " ,np.array(list_).dtype)

# printig data type of the array list
print("Datatype of array_list is : ",array_list.dtype)
```

```python
#Ans: 
# printig data type of every element of the list
for i in list_:
    print("Datatype of list is : " ,np.array(list_).dtype)

# printig data type of every element of the array list
for i in array_list:
    print("Datatype of array_list is : ",array_list.dtype)
    
```

```python
#Ans:
import numpy as np

list_ = [ '1' , '2' , '3' , '4' , '5' ]

array_list = np.array(object = list_, dtype = int)

# printig data type of every element of the list
for i in list_:
    print("Datatype of list is : " ,np.array(list_).dtype)

# printig data type of every element of the array list
for i in array_list:
    print("Datatype of array_list is : ",array_list.dtype)
```
